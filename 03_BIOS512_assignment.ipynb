Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width = 5, repr.plot.height = 5)
p = ggplot(penguins, aes(x = body_mass_g)) + 
        geom_histogram(na.rm = TRUE, bins = 30, fill = "Steelblue", color = "Steelblue", alpha = 0.7) + 
        facet_wrap(~year, ncol = 1) + 
        labs(x = "Body mass in grams", y = "Number of penguins", title = "Histograms of penguin body mass for measurements taken from 2007-2009") + 
        scale_x_continuous(labels = comma_format(), breaks = seq(2500, 6500, by = 1000)) + 
        scale_y_continuous(breaks = seq(0, 7.5, by = 2.5)) +
        theme_minimal() + 
        theme(
            plot.title = element_text(size = 8),
            axis.title.x = element_text(size = 9),
            axis.title.y = element_text(size = 9),
            strip.text = element_text(hjust = 0.925, face = "bold", colour = "grey55")
        ) 
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width = 4, repr.plot.height = 4)
p = ggplot(penguins, aes(x = body_mass_g, y = sex)) +
    geom_point(na.rm = TRUE, position = position_jitter(height = 0.25, width = 0), size = 1, alpha = 0.35) + 
    facet_wrap(~species, ncol = 1) +  
    scale_x_continuous(labels = comma_format(), breaks = seq(3500, 5500, by = 1000)) + 
    labs(x = "Body mass in grams") + 
    theme_minimal() +
    theme(
            axis.title.x = element_text(size = 9),
            axis.title.y = element_blank(),
            strip.text = element_text(hjust = 0.05, face = "bold", colour = "grey30", size = 8)
        ) 
p
```
